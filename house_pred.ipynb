```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data_h = pd.read_excel('/content/House price data .xlsx')
```

```python
data_h.shape
```

```python
data_h.info()
```

```python
data_h.head(10)
```

```python
X=data_h[['number of bedrooms','number of bathrooms','Postal Code','living area','lot area']]
Y=data_h['Price']
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(y_pred)
```

```python
acc=model.score(X_test,Y_test)
print(acc)
```

```python
new_pred =pd.DataFrame({'number of bedrooms':[2],'number of bathrooms':[2.5],'Postal Code':[98008],'living area':[600],'lot area':[600]})
```

```python
pred_price=model.predict(new_pred)
```
